# Dataframe con los dos datasets

Primero hacemos el dataset con las fotos de la carpeta 'limpieza' que ya estan en blanco y negro, recortadas y transformadas a un tamaño de 128x128

```python
cropped_nopersonas = getDataset('./Imagenes/limpieza/*.png')
```

```python
len(cropped_nopersonas)
```

```python
df_nopersona = pd.DataFrame(cropped_nopersonas)
```

```python
df_nopersona = df_nopersona.T
df_nopersona.head()
```

```python
df_nopersona = df_nopersona.rename(columns={0:'array'})
```

```python
df_nopersona = df_nopersona.reset_index()
```

```python
df_nopersona = df_nopersona.rename(columns={'index':'filename'})
```

```python
df_nopersona['persona'] = 0
```

```python
df_nopersona
```

Hacemos lo mismo con las fotos de la carpeta 'cropped' que contienen fotos de trozos de persona en blanco y negro y de tamaño 128x128

```python
cropped_personas = getDataset('./Imagenes/cropped/*.png')
```

```python
len(cropped_personas)
```

```python
df_personas = pd.DataFrame(cropped_personas)
```

```python
df_personas = df_personas.T
```

```python
df_personas = df_personas.rename(columns={0:'array'})
```

```python
df_personas = df_personas.reset_index()
```

```python
df_personas = df_personas.rename(columns={'index':'filename'})
```

```python
df_personas['persona'] = 1
```

```python
df_personas
```

Los concateno para solo tener uno:

```python
data = pd.concat([df_nopersona, df_personas], axis=0)
```

```python
data
```

```python
data.isnull().sum()
```

```python
data = data.reset_index(drop=True)
```
